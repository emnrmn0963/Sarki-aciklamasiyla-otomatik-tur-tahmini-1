```python
from collections import Counter 
import numpy as np 
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
# lemmatization işlemi sonrası Zipf grafiği
```

```python
dflem=pd.read_csv(r"C:\Users\lenovo\Desktop\metin_verileri_ile_dogal_dil_isleme\lemmatized_sentences.csv",header=None)
```

```python
sentences=dflem[0].dropna()
sentences=sentences[sentences.apply(lambda x:isinstance(x,str))]
```

```python
tokens=[word_tokenize(sentence) for sentence in sentences]
```

```python
flat_tokens=[token for sentence in tokens for token in sentence]
```

```python
word_freq=Counter(flat_tokens)
```

```python
sorted_freqs=sorted(word_freq.values(),reverse=True)
```

```python
ranks=range(1,len(sorted_freqs)+1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.log(ranks),np.log(sorted_freqs),marker="o",linestyle="None")
plt.xlabel("Log(Rank)")
plt.ylabel("Log(Frequency)")
plt.title("Log-Log Kelime Frekansı Dağılımı")
plt.grid(True)
plt.show()
```
